```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x.ndim) # 1
print(x.shape) # (3, )

# element-wise operation
print (x + y) 
print (x * y)

# broadcasting
print(x * 3)
a = np.array([[1, 2], [3, 4]])
b = np.array([10, 20])
print(a * b)

# 내적과 행렬곱
## 내적
print(np.dot(x, y))
## 행렬곱
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.matmul(A, B))
```
