```python
import numpy as np

class Affine :
	def __init__(self, W, b) :
		self.params = [W, b]

	def forward(self, x) :
		W, b = self.params
		return (np.matmul(x, W) + b)
	
class Sigmoid :
	def __init__(self) :
		self.params = []
	
	def forward(self, x) :
		return (1 / (1 + np.exp(-x)))

class TwoLayerNet :
	def __init__(self, input, hidden, output) :
		I, H, O = input, hidden, output
		W1 = np.random.randn(I, H) # hidden Layer Weight
		b1 = np.random.randn(1, H) # hidden Layer Bias
		W2 = np.random.randn(H, O)
		b2 = np.random.randn(1, O)

		# 계층들을 하나의 리스트로 만들어서 사용하기 쉽도록 구성
		self.layers = [
			Affine(W1, b1),
			Sigmoid(),
			Affine(W2, b2)
		]

		# 각 계층들의 매개 변수를 한 곳에 모아둠 : 역전파시 업데이트가 쉬워진다.
		self.params = []
		for layer in self.layers :
			self.params.append(layer.params) # <=> self.params += layer.params

	def predict(self, x) :
		for layer in self.layers :
			x = layer.forward(x)
		return (x)

batch_size = 3
x = np.random.randn(batch_size, 2) # input data
model = TwoLayerNet(2, 4, 3)
s = model.predict(x)

print(s)
```
