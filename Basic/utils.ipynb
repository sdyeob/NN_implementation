```python
# Back-Prop시 도움을 주는 여러 Utility Functions

# 1. Repeat function
import numpy as np

D, N = 8, 7
x = np.random.randn(1, D) # Input
y = np.repeat(x, N, axis=0) # 1,D의 x를 Axis=0방향으로 N개 생성 -> y.shape = (N, D)

dy = np.random.randn(N, D) # dL/dy
dx = np.sum(dy, axis=0, keepdims=True) # sum(dL/dy, axis=0)

# 2. Sum node : 덧셈 노드 확장
x = np.random.randn(N, D)
y = np.sum(x, axis=0, keepdims=True)

dy = np.random.randn(1, D) # dL/dy
dx = np.repeat(dy, N, axis=0)
```

```python
# 3. Matmul node
class MatMul :
	def __init__(self, W) :
		self.params = [W]
		self.grads = [np.zeros_like(W)]
		self.x = None

	def forward(self, x) :
		self.x = x
		W = self.params
		out = np.matmul(x, W)
		return out
	
	def backward(self, dout) :
		W = self.params
		dx = np.matmul(dout, W.T)
		dW = np.matmul(self.x.T, dout)
		self.grads[0][...] = dW # 생략기호 [...]를 이용하면, 메모리에 직접 Write

		return dx
```
